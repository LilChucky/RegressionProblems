```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as opt
```

```python
#1 Generating random numbers
inputValues = np.random.uniform(-4,4, 150)
```

```python
#2 A custom function
def myCustFunc(x,l1,l2):
  return l1/np.exp(x)+l2*np.sin(x)
```

```python
# 3 Setting a train set, i guess !?
trainSet = inputValues[:100]
# testSet = inputValues[100:]
trainSetOutput = myCustFunc(trainSet,0.65,7)
```

```python
# 4 Hello Plot!!! (☞ﾟヮﾟ)☞
plt.plot(trainSet,trainSetOutput,'.',label='Actual')
plt.legend(loc='best')
plt.xlabel ('input')
plt.ylabel ('output')
plt.title('First Graph')

plt.show()
```

```python
# 5 Loud Data (☞ﾟOﾟ)☞
noisyOutputValues = trainSetOutput+\
 np.random.lognormal(0., 0.85, 100)
```

```python
# 7 A fourth degree polynomial function :)
def poly4thGegree(x,a,b,c,d,e):
  return a*(x**4)+b*(x**3)+c*(x**2)+d*x+e
```

```python
# 6 My first time using the curve_fitting function (:
# fitting on original data
bestValsPolyModel, CoVarPolyModel =\
 opt.curve_fit(poly4thGegree,trainSet,trainSetOutput)
print('Polynomial model estimated parameters (on original data) are:',\
      '{:.2f}'.format(bestValsPolyModel[0]),\
      '{:.2f}'.format(bestValsPolyModel[1]),\
      '{:.2f}'.format(bestValsPolyModel[2]),\
      '{:.2f}'.format(bestValsPolyModel[3]),\
      '{:.2f}'.format(bestValsPolyModel[4]))
#
#
# fitting on original data
bestValsPolyModel, CoVarPolyModel =\
 opt.curve_fit(poly4thGegree,trainSet,noisyOutputValues)
print('Polynomial model estimated parameters (on noisy data) are:',\
      '{:.2f}'.format(bestValsPolyModel[0]),\
      '{:.2f}'.format(bestValsPolyModel[1]),\
      '{:.2f}'.format(bestValsPolyModel[2]),\
      '{:.2f}'.format(bestValsPolyModel[3]),\
      '{:.2f}'.format(bestValsPolyModel[4]))
```

```python
# 8 Value Prediction
polyModelPredictedOutcomes =\
 poly4thGegree(trainSet, bestValsPolyModel[0], bestValsPolyModel[1],\
               bestValsPolyModel[2], bestValsPolyModel[3], bestValsPolyModel[4] )
```

```python
# 9 Another graph
plt.figure(figsize= (20,10))
plt.plot(trainSet,trainSetOutput,'*r',label='Actual Data')
plt.plot(trainSet,noisyOutputValues,'.c',label='Noisy Data')
plt.plot(trainSet,polyModelPredictedOutcomes,'og',label = 'Predicted')

plt.legend(loc='best')
plt.ylabel ('input')
plt.xlabel ('output')
plt.title('Polynomial Graph')
plt.show()
```
